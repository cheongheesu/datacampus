```python
import pandas as pd

# 1. JSON 파일 불러오기
json_path = '/content/drive/MyDrive/상거래(쇼핑).json'   # 실제 JSON 파일 경로로 수정해주세요
df22 = pd.read_json(json_path)

# "data" 컬럼의 각 값에서 "body" 항목의 "utterance"를 추출
def extract_utterances(data_item):
    utterances = [item['utterance'] for item in data_item['body']]
    return utterances

df22['utterances'] = df22['data'].apply(extract_utterances)
```
